```python
import scanpy as sc 
import matplotlib.pyplot as plt 
sc.settings.figdir = "..results/figures"
```

```python
adata=sc.read_h5ad('../data/processed/reduced.h5ad')
```

```python
#cluster 
sc.tl.leiden(adata,resolution=0.5)
```

```python
#visualize 

sc.pl.umap(adata,color='leiden')
```

```python
for res in [0.2,0.5,0.8,1.0]:
    sc.tl.leiden(adata,resolution=res,key_added=f'leiden_{res}')
```

```python
sc.pl.umap(
    adata,
    color=['leiden_0.2',
           'leiden_0.5',
           'leiden_0.8',
           'leiden_1.0'],
    save='_resolution_comparison.png'
)
```

```python
adata.write("../data/processed/nsclc_clustered.h5ad") 
```
