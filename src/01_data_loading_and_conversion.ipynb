```python
import os
import gzip
import shutil
import urllib.request
```

## Download files

```python
base_url = "https://ftp.ncbi.nlm.nih.gov/geo/series/GSE131nnn/GSE131907/suppl/"

files = [
    "GSE131907_Lung_Cancer_raw_UMI_matrix.txt.gz",
]

for file in files:
    url = base_url + file
    output_path = f"../data/raw/{file}"
    

    print(f"Downloading {file}...")
    urllib.request.urlretrieve(url, output_path)

print("Download completed.")
```

```python
import scanpy as sc

adata = sc.read_text(
    "../data/raw/GSE131907_Lung_Cancer_raw_UMI_matrix.txt.gz"
).T

adata
```

```python
from scipy.sparse import csr_matrix
adata.X = csr_matrix(adata.X)  
```

```python
adata.write('../data/raw/GSE131907_Lung_Cancer_raw.h5ad',compression='lzf')
```
